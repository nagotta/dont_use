## Ë™≤È°å1
ÂæÆÂàÜÊñπÁ®ãÂºè ùëëùë•/ùëëùë°=ùë• „Å´„Å§„ÅÑ„Å¶„ÄÅ„Ç™„Ç§„É©„ÉºÊ≥ï„Å®„É´„É≥„Ç≤„ÇØ„ÉÉ„ÇøÊ≥ï„Å´„Çà„ÇäÊï∞ÂÄ§Ë®àÁÆó„Åó„ÄÅ
 „Åù„ÅÆ„Ç∞„É©„Éï„ÇíÊèè„Åë„ÄÇÂàùÊúüÊù°‰ª∂„ÅØ ùë•0 = 1„Å®„Åô„Çã„ÄÇ

```python
import matplotlib.pyplot as plt
# dx/dt = f(x) = x
def f(x, t):
    return x
step = 0.05 # Âàª„ÅøÂπÖ
# „Ç™„Ç§„É©„ÉºÊ≥ï
x = 1 # x„ÅÆÂàùÊúüÂÄ§
t = 0 # t„ÅÆÂàùÊúüÂÄ§
ts_euler = [] # xÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà
xs_euler = [] # yÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà
for i in range(1, 100):
# „Ç™„Ç§„É©„ÉºÊ≥ï„Å´„Çà„ÇãÊõ¥Êñ∞Âºè
    x = x + f(x, t) * step
    t += step # t„ÅÆÂÄ§„ÇíÊõ¥Êñ∞
    ts_euler.append(t) # xÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà„Å´ËøΩÂä†
    xs_euler.append(x) # yÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà„Å´ËøΩÂä†
# „É´„É≥„Ç≤„ÇØ„ÉÉ„ÇøÊ≥ï
x = 1 # x„ÅÆÂàùÊúüÂÄ§
t = 0 # t„ÅÆÂàùÊúüÂÄ§
ts_rk = [] # xÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà
xs_rk = [] # yÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà
for i in range(1, 100):
    # „É´„É≥„Ç≤„ÇØ„ÉÉ„ÇøÊ≥ï„Å´„Çà„ÇãÊõ¥Êñ∞Âºè
    k1 = f(x, t)
    k2 = f(x + k1 * step / 2, t + step / 2)
    k3 = f(x + k2 * step / 2, t + step / 2)
    k4 = f(x + k3 * step / 2, t + step)
    x = x + (k1 + 2 * k2 + 2 * k3 + k4) * step / 6
    t += step # t„ÅÆÂÄ§„ÇíÊõ¥Êñ∞
    ts_rk.append(t) # xÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà„Å´ËøΩÂä†
    xs_rk.append(x) # yÂ∫ßÊ®ô„ÅÆ„É™„Çπ„Éà„Å´ËøΩÂä†
# Ëß£ÊûêËß£„ÅÆ„Ç∞„É©„Éï
import numpy as np
ts_exp = np.arange(1,5, 0.01)
xs_exp = np.exp(ts_exp)
# „Ç∞„É©„Éï„ÅÆÊèèÁîª
plt.plot(ts_euler, xs_euler, label="Euler method")
plt.plot(ts_rk, xs_rk, label="Runge‚ÄìKutta method")
plt.plot(ts_exp, xs_exp, label="$e^t$")
plt.legend() # Âá°‰æã„ÅÆË°®Á§∫
plt.show() # „Ç∞„É©„Éï„ÅÆË°®Á§∫
```

```python
import numpy as np
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import math

# dx/dt = cos(t)x
def func1(x, t):
    return math.cos(t)*x
def func2(t, x):
    return math.cos(t)*x
# odeint
# t„ÅÆÁØÑÂõ≤(1<=t<10„ÅÆÁØÑÂõ≤„Åß„ÄÅ0.01ÈñìÈöî„ÅÆ„É™„Çπ„Éà„ÇíËøî„ÅôÔºâ
ts_ode = np.arange(0, 10, 0.01)
x = 1 # x„ÅÆÂàùÊúüÂÄ§
xs_ode = odeint(func1, x, ts_ode)
# „Ç∞„É©„Éï„ÅÆÊèèÁîª
plt.plot(ts_ode, xs_ode, label="odeint", linestyle='--')
# solve_ivp
t_span = (0, 10)
t_eval = np.linspace(*t_span, 100)
sol = solve_ivp(func2, t_span, [1], t_eval=t_eval)
plt.plot(sol.t, sol.y[0], label="solve_ivp", linestyle=':')
# „Ç∞„É©„Éï„ÅÆÊèèÁîª
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
```
