‚Ä¢ ùëÜ0 = 100„ÄÅùêº0 = 1„ÄÅùëÖ0 = 0„ÄÅùúÜ = 0.0004„ÄÅ ùõæ = 0.2„ÅÆÂ†¥Âêà„ÅßÂÆüË°å„Åó„ÄÅÁµêÊûú„Åå„Çπ„É©„Ç§„Éâ„Å®Âêå„Åò„Å´„Å™„Çã„Åì„Å®„ÇíÁ¢∫Ë™ç„Åõ„Çà„ÄÇ

```python
# ùëÜ0 = 1000„ÄÅùêº0 = 1„ÄÅùëÖ0 = 0„ÄÅùúÜ = 0.0004„ÄÅ ùõæ = 0.2

from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def f(vars, t, Œª, Œ≥):
    S,I,R = vars
    return -Œª*S*I, Œª*S*I-Œ≥*I, Œ≥*I

S0 = 1000       # Susceptible-ÊÑüÂèóÊÄß‰∫∫Âè£
I0 = 1          # Infections-ÊÑüÊüì‰∫∫Âè£
R0 = 0          # Removed-ÈöîÈõ¢„Åï„Çå„Åü‰∫∫Âè£
Œª = 0.0004      # ‰øÇÊï∞Œª-ÊÑüÊüìÁéá
Œ≥ = 0.2         # ‰øÇÊï∞Œ≥-ÂõûÂæ©Áéá„ÇÑÈöîÈõ¢Áéá
S,I,R = S0,I0,R0

ts_ode = np.arange(0.0, 100.0, 0.01)
S_ode = odeint(f,[S,I,R],ts_ode,args=(Œª,Œ≥,))
plt.plot(ts_ode,S_ode[:,0],label="Susceptible",linestyle='-')
plt.plot(ts_ode,S_ode[:,1],label="Infections",linestyle='-')
plt.plot(ts_ode,S_ode[:,2],label="Removed",linestyle='-')
plt.legend()
plt.show()
```

## 3. [ÈÅ∏Êäû] Ë¨õÁæ©„ÅßÂ≠¶„Çì„Å†‰ª•Â§ñ„ÅÆÂæÆÂàÜÊñπÁ®ãÂºè„ÇíÁî®„ÅÑ„Åü„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥„ÇíÂÆüË£Ö„Åó„Å¶ÁµêÊûú„ÇíË™¨Êòé„Åõ„Çà„ÄÇ

### ÂçòÊåØ„ÇäÂ≠ê„ÅÆÈÅãÂãï
- ÂçòÊåØ„ÇäÂ≠ê„ÅÆÈÅãÂãï„ÅÆÂæÆÂàÜÊñπÁ®ãÂºè„ÇíÁî®„ÅÑ„Åü„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥„ÇíË°å„Å£„Åü.
- ÊåØ„ÇäÂ≠ê„ÅÆ‰ΩçÁΩÆ„ÇíŒ∏„ÄÅÊåØ„ÇäÂ≠ê„ÅÆÈÄüÂ∫¶„Çíœâ„Å®„Åó„ÅüÂ†¥Âêà
  - dŒ∏/dt = œâ
  - dœâ/dt = -g*sin(Œ∏)/L   (g„ÅØÈáçÂäõÂä†ÈÄüÂ∫¶,L„ÅØÊåØ„ÇäÂ≠ê„ÅÆÈï∑„Åï)

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def f(vars, t, g, L):
    Œ∏,œâ = vars
    return œâ, -(g/L)*np.sin(Œ∏)

Œ∏0 = np.pi/4     # ÂàùÊúüËßíÂ∫¶: œÄ/4
œâ0 = 0           # ÂàùÈÄüÂ∫¶: 0
Œ∏,œâ = Œ∏0,œâ0
g = 9.80665      # ÈáçÂäõÂä†ÈÄüÂ∫¶: 9.80665
L = 1.0          # ÊåØ„ÇäÂ≠ê„ÅÆÈï∑„Åï: 1

ts_ode = np.arange(0.0, 15.0, 0.01)
S_ode = odeint(f,[Œ∏,œâ],ts_ode,args=(g,L,))
plt.plot(ts_ode,S_ode[:,0],label="Angle: Œ∏",linestyle='-')
plt.plot(ts_ode,S_ode[:,1],label="Angular V: œâ",linestyle='-')
plt.xlabel("Time")
plt.ylabel("A")
plt.legend()
plt.show()
print("MaxŒ∏:",np.max(S_ode))
```
