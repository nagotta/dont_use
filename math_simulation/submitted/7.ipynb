1. „Ç¶„Çµ„ÇÆ„Å®„Ç≠„ÉÑ„Éç„ÅÆÁîüÊÖãÁ≥ª„ÅÆ„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥„ÅÆ„Éó„É≠„Ç∞„É©„É†„Çí‰ΩúÊàê„Åõ„Çà„ÄÇ
ùêµùëü = 1Ôºåùëé = 0.1Ôºåùëè = 0.02Ôºåùê∑ùëì = 1ÔºåùëÖ 0 = 3 Ôºåùêπ 0 = 1„ÅÆÂ†¥Âêà„Åß„Ç∑„Éü„É•„É¨„Éº„Ç∑„Éß„É≥„ÇíÂÆüË°å„Åõ„Çà„ÄÇ

```python
from scipy.integrate import odeint
Br = 1
a = 0.1
b = 0.02
Df = 1
R0 = 3
F0 = 1
R = R0
F = F0

def func1(vars, t, Br, a, b, Df):
    R,F = vars
    return (Br-a*F)*R, (b*R-Df)*F

import matplotlib.pyplot as plt
import numpy as np
ts_ode = np.arange(0.0, 14.0, 0.01)
Rs_ode = odeint(func1,[R,F],ts_ode,args=(Br,a,b,Df,))
plt.plot(ts_ode,Rs_ode[:,0],label="Rabbit",linestyle='--')
plt.plot(ts_ode,Rs_ode[:,1],label="Fox",linestyle='--')

plt.legend()
plt.show()
```

4. [‰ªªÊÑè] 3Á®Æ„ÅÆÈñ¢‰øÇ„Åß„ÅØ„Å©„ÅÜ„Å™„Çã„Å†„Çç„ÅÜ„Åã„ÄÇ„Åü„Å®„Åà„Å∞„ÄÅ„Ç¶„Çµ„ÇÆ„Å®„Ç≠„ÉÑ„Éç„Å´ Âä†„Åà„Å¶„ÄÅ„Ç≠„ÉÑ„Éç„ÇíÈ£ü„Åπ„Çã„Ç™„Ç™„Ç´„Éü„Åå„ÅÑ„ÅüÂ†¥Âêà„ÄÅ3Á®Æ„ÅÆÂÄã‰ΩìÊï∞„ÅÆÂ§âÂåñ„ÅØ„Å© „ÅÆ„Çà„ÅÜ„Å´„Å™„Çã„Åã„Éó„É≠„Ç∞„É©„É†„ÇíÊîπËâØ„Åó„ÄÅË™¨Êòé„Åõ

W0„Å®W„ÇíËøΩÂä†
Bw:Áãº„ÅÆÂá∫ÁîüÁéá
Df -> Dw:Áãº„ÅÆÊ≠ª‰∫°Áéá
c,d:ÊØî‰æãÂÆöÊï∞
W(t+Œît) - W(t) = (Bw - Dw)W(t)Œît
Áãê„ÅÆÊ≠ª‰∫°Áéá Df = cW(t)„ÄÅÁãº„ÅÆÂá∫ÁîüÁéáBw = dF(t)„ÇíÁî®„ÅÑ„Å¶‰ª•‰∏ã„ÅÆÂºè„ÇíË°®„Åô
F(t+Œît) = (1+(bR(t) - cW(t))Œît)F(t)
W(t+Œît) = (1+(dF(t) - Dw)Œît)W(t)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

Br = 1
a = 0.1
b = 0.02
c = 0.2
d = 0.1
Dw = 1
R0 = 3
F0 = 2
W0 = 1
R = R0
F = F0
W = W0

def func1(vars, t, Br, a, b, c, d, Dw):
    R,F,W = vars
    return (Br-a*F)*R, (b*R-c*W)*F,(d*F-Dw)*W

ts_ode = np.arange(0.0, 30.0, 0.01)
Rs_ode = odeint(func1,[R0,F0,W0],ts_ode,args=(Br,a,b,c,d,Dw,))
plt.plot(ts_ode,Rs_ode[:,0],label="Rabbit",linestyle='-')
plt.plot(ts_ode,Rs_ode[:,1],label="Fox",linestyle='-')
plt.plot(ts_ode,Rs_ode[:,2],label="Wolf",linestyle='-')

plt.legend()
plt.show()
```
