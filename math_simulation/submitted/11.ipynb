```python
import random
# エージェントのクラス
class Agent:
    R = 0.2 # エージェントの近傍
    # エージェントの初期状態の設定
    def __init__(self, category):
        self.category = category
        self.x = 0
        self.y = 1
    # エージェントの状態を出力
    def __str__(self):
        return f'x={self.x}¥ty={self.y}'
    # エージェントの次の状態を計算
    def next(self, agents):
        if self.category == 0:
            # エージェント カテゴリ0の動作
            for agent in agents:
                if agent.category == 1:
                    # カテゴリ1のエージェントとの距離を調べる
                    my_x = self.x
                    my_y = self.y
                    you_x = agent.x
                    you_y = agent.y
                    if (my_x - you_x) ** 2 + (my_y - you_y) ** 2 < Agent.R:
                        # 隣接してカテゴリ1のエージェントが存在する
                        self.category = 1
                        break # ループを抜ける
            # 自身のカテゴリが変わらなかった場合
            if self.category == 0:
                # 次の座標をランダムに決定
                self.x += (random.random() - 0.5) * 2
                self.y += (random.random() - 0.5) * 2
        elif self.category == 1:
            # エージェント カテゴリ1の動作
            # 次の座標をランダムに決定
            self.x += 0.1
            self.y += 0.1
        else:
            # 合致するカテゴリがない場合
            print(f'ERROR: カテゴリ {self.category} がありません')
```

```python
import matplotlib.pyplot as plt
if __name__ == '__main__':
    STEP_NUM = 100 # 時間
    N = 30 # エージェントの個数
    # カテゴリ0のエージェントのx, y座標のリスト
    xlist0 = []
    ylist0 = []
    # カテゴリ1のエージェントのx, y座標のリスト
    xlist1 = []
    ylist1 = []
    # エージェントの配列（Agent N個）
    agents = [Agent(0) for i in range(N)]
    # カテゴリ1のエージェントを設定
    agents[0].category = 1
    agents[0].x = -2
    agents[0].y = -2

# 時間分ループ
    for t in range(STEP_NUM):
        plt.clf() # グラフをクリア
        for agent in agents:
            # エージェントの次の状態を計算
            agent.next(agents)
            # エージェントのx座標とy座標を取得（描画用）
            if agent.category == 0:
                xlist0.append(agent.x)
                ylist0.append(agent.y)
            elif agent.category == 1:
                xlist1.append(agent.x)
                ylist1.append(agent.y)
        # シミュレーショの領域（40x40ピクセル）を設定
        plt.axis([-20, 20, -20, 20])
        # エージェントをグラフにプロット
        plt.plot(xlist0, ylist0, '.')
        plt.plot(xlist1, ylist1, '+')
        # 可視化のため一時停止
        plt.pause(0.1)
        # 座標をクリア
        xlist0.clear()
        ylist0.clear()
        xlist1.clear()
        ylist1.clear()
        plt.savefig(f'img{t:02}.png')
```
